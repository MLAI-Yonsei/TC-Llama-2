```python
taskXX_data = pd.DataFrame()

# Generate 'context' column



taskXX_data['input'] = '과제명: ' + rawdata['NTIS_A_F_014'] +"\n\n" \
                      + '연구목표: ' + rawdata['NTIS_A_F_076']+"\n\n"\
                      + '연구내용: ' + rawdata['NTIS_A_F_080']+"\n\n"\
                      + '기대효과: ' + rawdata['NTIS_A_F_084']+"\n\n"


# Generate 'response' column
taskXX_data['output'] ='사업화명: ' + rawdata['NTIS_C_F_014']+"\n\n"\
                      + '사업화내용:' + rawdata['NTIS_C_F_018'] +"\n\n"\

taskXX_data['instruction'] = "당신은 연구과제를 사업화하는 전문가입니다. 다음 input(과제명, 연구목표, 연구내용, 기대효과)을 기반으로 가능한 사업 아이디어를 생성하시오. 사업 아이디어(business idea)는 연구과제를 기반으로 한 새로운 사업 아이디어를 설명합니다. 그리고 사업 아이디어에 따른, 구체적인 제품 7개를 생성하시오. 제품은 연구과제와 사업아이디어에 따른 구체적 제품명으로, 실제로 소비자에게 판매될 수 있는 형태여야 하며, 추론에 따라 합리적이며 다양해야합니다. 유의사항: 입력된 마지막 문장은 불완전할 수 있으므로, 제공된 정보로 최선의 답변을 해주세요. 답변은 모두 연구과제와 관련되어야 하며, 간결해야 합니다. 문장의 첫시작은 '제품명1 : ' 으로 시작하십시오."


taskXX_data = taskXX_data[['instruction','input','output']]

taskXX_data.head()
```

```python
model_name = "daryl149/llama-2-7b-chat-hf"#"NousResearch/Llama-2-7b-chat-hf"

from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
```

```python
(taskXX_data['instruction']+taskXX_data['input'])
```

```python
taskXX_data['encoded'] = (taskXX_data['instruction']+taskXX_data['input']).apply(lambda x : len(tokenizer.encode(x)))
```

```python
taskXX_data[taskXX_data['encoded'] < 1800]
```

```python
taskXX_datas = taskXX_data[taskXX_data['encoded'] < 1800].sample(n=50, random_state=42) #랜덤하게 샘플링  # llama2의 토크나이저를 사용하여 모든 토큰이 전부 input값으로 들어갈 수 있는 샘플들만 선별
```

```python
taskXX_datas = taskXX_datas.reset_index(drop=True).astype("str")
```

```python
taskXX_datas
```

```python
# 데이터프레임을 딕셔너리로 변환하여 저장할 리스트 생성
data_list1 = []

# 데이터프레임의 각 행에 대해 딕셔너리 생성 후 리스트에 추가
for _, row in taskXX_datas.iterrows():
    data_dict1 = {
        'instruction' : row['instruction'],
        'input': row['input'] ,
        'output': row['output']    }
    data_list1.append(data_dict1)

# 딕셔너리 리스트를 JSON 파일로 저장
with open(file_path + '/KISTI/NTISdata_generating/kisti_taskXX.json', 'w',encoding='utf-8') as f:
    json.dump(data_list1, f, ensure_ascii=False)
```

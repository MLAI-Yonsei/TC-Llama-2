## Task 2 : 이어질 카테고리 설명 예측

### 예시  
**instruction**  
다음은 분류체계와 분류체계에 대한 상세 설명 일부입니다. 이어지는 설명을 쓰세요.

**input**  
산동식물및동식물성생산품 ; 생명이 있는 야생과 재배성의 식물과 동물, 생명이 있는 식물들과 그들의

**output**  
씨앗, 생명이 있는 식물과 동물의 보호, 수용, 사육에 필요한 재료와 설비

```python
#설명을 단어 기준으로 쪼개서 절반을 Input, 절반을 output
def split_words(sentence):
  return sentence.split( )

def group_words(words):
  half_length = len(words) // 2
  first_half_words = words[:half_length]
  second_half_words = words[half_length:]
  first_half_words = ' '.join(first_half_words)
  second_half_words = ' '.join(second_half_words)
  return first_half_words, second_half_words

def get_first_half_words(sentence):
  return group_words(split_words(sentence))[0]

def get_second_half_words(sentence):
  return group_words(split_words(sentence))[1]

appended1['first_half_words'] = appended1['Definition'].apply(get_first_half_words)
appended1['second_half_words'] = appended1['Definition'].apply(get_second_half_words)

appended1.head()
```

```python
len(rawdata)
```

```python
task2_data = pd.DataFrame()
# Generate 'context' column
task2_data['context'] = "Taxonomy: " + appended1['Title']\
                        +"\n"+ "Definition : " + appended1['first_half_words']

# Generate 'response' column
task2_data['response'] = appended1['second_half_words']


task2_data['instruction'] = "The following is a taxonomy and a partial description of the taxonomy. Read the following partial definition description of the taxonomy and provide a coherent completion that accurately defines the taxonomy term. "
task2_data['category'] = "task2"

task2_data['index'] = task2_data.index

task2_data = task2_data[['instruction','context','response','category','index']]

task2_data.tail()
```

## Task 3 : 상,하위 카테고리 예측



### 예시  
물품목록번호는 2자리씩 4단계 계층구조로 이루어진 8자리 숫자이다.

**instruction**  
다음은 물품분류번호와 해당하는 분류체계입니다.이 분류체계의 상위 분류체계를 예측하시오.   

**input**  
물품분류번호는 5126 입니다. 분류체계는 신경계감각기관용의약품 입니다.   
                
**output**  
대분류: 의약품

```python
#ㅇㅇ : lv1(대분류)  , ㅇㅇ : lv2(중분류) ,ㅇㅇ : lv3(소분류), ㅇㅇ-ㅇㅇ : 품명 - 세부번호
#계층 데이터프레임 만들기
lv_data = pd.DataFrame()

lv_data['thngClsfcNoNm']=rawdata['thngClsfcNoNm']
lv_data['thngClsfcNo']=rawdata['thngClsfcNo']
lv_data['대분류_No'] = rawdata['thngClsfcNo'].astype(str).str[:2]
lv_data['중분류_No'] = rawdata['thngClsfcNo'].astype(str).str[:4]
lv_data['소분류_No'] = rawdata['thngClsfcNo'].astype(str).str[:6]
lv_data['품명_No'] = rawdata['thngClsfcNo'].astype(str).str[:8]
lv_data['세부번호_No'] = rawdata['thngClsfcNo'].astype(str).str[:10]


lv_data['대분류_NoNm'] = lv_data['대분류_No'].map(lv_data.groupby('대분류_No')['thngClsfcNoNm'].first())
lv_data['중분류_NoNm'] = lv_data['중분류_No'].map(lv_data.groupby('중분류_No')['thngClsfcNoNm'].first())
lv_data['소분류_NoNm'] = lv_data['소분류_No'].map(lv_data.groupby('소분류_No')['thngClsfcNoNm'].first())
lv_data['품명_NoNm'] = lv_data['품명_No'].map(lv_data.groupby('품명_No')['thngClsfcNoNm'].first())

lv_data['index'] = lv_data.index

lv_data = lv_data[['thngClsfcNo','thngClsfcNoNm','대분류_No','중분류_No','소분류_No','품명_No','세부번호_No',
                   '대분류_NoNm','중분류_NoNm','소분류_NoNm','품명_NoNm', 'index']]

lv_data.head(100)
```

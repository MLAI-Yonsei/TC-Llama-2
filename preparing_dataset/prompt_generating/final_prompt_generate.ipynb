## train-test split function

```python
split_ratio = 0.7
seed = 42
```

```python
import pandas as pd
import numpy as np


def train_test_split_by_group(group, train_ratio=0.7, seed=42):
    np.random.seed(seed)
    shuffled_indices = np.random.permutation(len(group))
    train_size = int(round(len(group) * train_ratio))
    print(group['data'].tolist()[0],group['category'].tolist()[0],train_size)
    train_indices = shuffled_indices[:train_size]
    test_indices = shuffled_indices[train_size:]
    return group.iloc[train_indices], group.iloc[test_indices]

```

```python
#NTIS_task2

def dataset_train_test_split(dataset,split_ratio=0.7):

  grouped = dataset.groupby(['data', 'category'])
  train, test = zip(*grouped.apply(lambda x: train_test_split_by_group(x, train_ratio=split_ratio, seed=seed)))

  # train_data와 test_data를 각각 하나의 데이터프레임으로 합치기
  trainset = pd.concat(train)
  testset = pd.concat(test)

  return trainset, testset
```

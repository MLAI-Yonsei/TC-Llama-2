## Task 4 : sentiment task (긍/부정 분류)

#### 별점데이터에 따라 긍정(별점 4,5점, 총 1452개) / 부정(별점 1,2,3점, 총 174)으로 분류

### 예시  
**instruction**  
다음은 제품에 대한 리뷰입니다. 해당 리뷰가 긍정적인지 부정적인지 분류하시오.  

**input**  
”두번째 구매입니다. 처음엔 마스크 얇고 힘이 없어서 말할 때 입에 달라붙으니까 별로였는데 다 써가니까 다시 구매하게 됐네요. 일단 말할 때 달라붙지만 그보다 장점이 많아요.안쪽이 부드러워서 얼굴에 편해요. 일반 덴탈마스크는 복불복인데 잘못 사면 안쪽 많이 거칠고 불편한데 이 마스크 부드러워서 좋아요. 그리고 절개선이 새로 되니까 얼굴 슬림해 보여요. 가로로 된 덴?마스크는 얼굴이 커보이는 효과 있어요.”

**output**  
긍정

```python
rawdata2[3].head(5)
```

```python
rawdata2[3].columns
```

```python
def rating_to_label(rating):
    if rating > 3:
        return "긍정"
    else:
        return "부정"

rawdata2[3]['sentiment'] = rawdata2[3]['rating'].apply(lambda x: rating_to_label(x))
```

```python
rawdata2[3].head(5)
```

```python
len(rawdata2[3])
```

```python
task4_data = pd.DataFrame()

# Generate 'context' column
task4_data['context'] = rawdata2[3]['content']

# Generate 'response' column
task4_data['response'] = rawdata2[3]['sentiment']



task4_data['instruction'] = "다음은 제품에 대한 리뷰입니다. 해당 리뷰가 긍정적인지 부정적인지 분류하시오."
task4_data['category'] = "task4"

task4_data['index'] = task4_data.index

task4_data['data'] = "danawa2-3"

task4_data['product_key'] = rawdata2[3]['pcode']


task4_data = task4_data[['instruction','context','response','category','index','data','product_key']]

task4_data.head()
```

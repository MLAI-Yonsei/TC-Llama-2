## Task 3 - 발명의 명칭을 보고 요소를 예측 + masking 된 것을 채워넣기

### 예시  
**instruction**  
다음은 발명의 명칭과 발명에 대한 설명의 일부입니다. 설명에 이어지는 문장을 완성하세요.

**input**  
발명의 명칭 : 선박의 액화가스 운용 시스템
발명에 대한 설명 : 선박의 액화가스 운용 시스템이 개시된다. 본 발명의 일 측면에 따르면, 액화가스가 저장되는 저장탱크; 상기 액화가스(LNG)를 기화시켜 제1 수요처로 공급하도록 기화기가 마련되는 제1 라인; 상기 저장탱크에서 발생된 증발가스(BOG)를 가열하여 제2 수요처로 공급하도록 가열기가 마련되는 제2 라인; 상기 제2

**output**  
라인으로부터 분기되어 제2 수요처로 증발가스를 공급하는 제3 라인; 및 상기 제3 라인 상에 마련되어 증발가스와 열교환하며 냉열을 제공받아 선박에 마련된 숙소(Accommodation) 및 조리실(Galley)로 냉방을 제공하는 공기조화기(AHU : Air Handling Unit);를 포함하는 선박의 액화가스 운용 시스템이 제공될 수 있다.

```python
#설명을 단어 기준으로 쪼개서 절반을 Input, 절반을 output
def split_words(sentence):
  return sentence.split( )

def group_words(words):
  half_length = len(words) // 2
  first_half_words = words[:half_length]
  second_half_words = words[half_length:]
  first_half_words = ' '.join(first_half_words)
  second_half_words = ' '.join(second_half_words)
  return first_half_words, second_half_words

def get_first_half_words(sentence):
  return group_words(split_words(sentence))[0]

def get_second_half_words(sentence):
  return group_words(split_words(sentence))[1]

rawdata['first_half_words'] = rawdata['요약'].apply(get_first_half_words)
rawdata['second_half_words'] = rawdata['요약'].apply(get_second_half_words)

rawdata.head()
```

```python
task3_data = pd.DataFrame()
# Generate 'context' column
task3_data['context'] = "발명의 명칭: " + rawdata['발명의 명칭']\
                        + " 발명에 대한 설명 : " + rawdata['first_half_words']

# Generate 'response' column
task3_data['response'] = rawdata['second_half_words']


task3_data['instruction'] = "다음은 발명의 명칭과 발명에 대한 설명의 일부입니다. 설명에 이어지는 문장을 완성하세요."
task3_data['category'] = "task3"

task3_data['index'] = task3_data.index

task3_data['data'] = "wipson"

task3_data = task3_data[['instruction','context','response','category','index','data']]

task3_data.head()
```

```python
appended_table = pd.concat([task1_data,task2_data,task3_data], ignore_index=True)
```

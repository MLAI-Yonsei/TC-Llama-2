## Task: 품목 -> 품명

### 예시
**instruction**
다음은 상품의 세부품목, 제조업체명, 상품모델명, . 해당하는 분류체계를 예측하시오.

**input**
생명이 있는 야생과 재배성의 식물과 동물, 생명이 있는 식물들과 그들의 씨앗, 생명이 있는 식물과 동물의 보호, 수용, 사육에 필요한 재료와 설비
세부품명은 퍼걸러이고 Col A 는 a 이고 Col B는 b 이다. 이때 품목을 맞추세요.”

**output**
산동식물및동식물성생산품

```python
#품목 column에 상품모델명까지는 담겨있어야함
rawdata1= rawdata1[rawdata1['품목'].str.count(',') >= 2]
```

```python
len(rawdata1)
```

```python
parsed_data_3 = []
parsed_data_4 = []

for index, row in rawdata1.iterrows():
    items = row['품목'].split(',')
    if len(items) == 3:
        parsed_data_3.append(row)
    elif len(items) >= 4:
        parsed_data_4.append(row)

# 분리된 데이터로 새로운 데이터프레임 생성
rawdata_3 = pd.DataFrame(parsed_data_3)
rawdata_4 = pd.DataFrame(parsed_data_4)
```

```python
rawdata_3.head()
```

```python
len(rawdata_3)
```

```python
len(rawdata_4)
```

```python
rawdata_4.head()
```

```python
# 데이터프레임 순회하면서 분할 작업 수행
parsed_data = []
parsed_data2 = []

for row in rawdata_3['품목']:
    items = row.split(',')
    세부품명 = items[0]
    제조업체명 = items[1]
    상품모델명 = items[2]
    result = f"상품의 세부품명은 {세부품명}"
    result2 = f"입니다. 제조업체명은{제조업체명}이고 상품모델명은{상품모델명} 입니다."

    parsed_data.append(result)
    parsed_data2.append(result2)

rawdata_3['parse1'] = parsed_data
rawdata_3['parse2'] = parsed_data2
#print(parsed_df)
```

```python
rawdata_3.head()
```

```python
parsed_data = []
parsed_data2 = []

for row in rawdata_4['품목']:
    items = row.split(',')
    세부품명 = items[0]
    제조업체명 = items[1]
    상품모델명 = items[2]
    부연설명 = ', '.join(items[3:])
    result = f"상품의 세부품명은 {세부품명}"
    result2 = f"입니다. 제조업체명은{제조업체명}이고 상품모델명은{상품모델명} 입니다. 상품 추가 정보는{부연설명} 입니다."

    parsed_data.append(result)
    parsed_data2.append(result2)

rawdata_4['parse1'] = parsed_data
rawdata_4['parse2'] = parsed_data2
```

```python
rawdata_4.head()
```

```python
#instruction #context #response #category

task1_1data = pd.DataFrame()

# Generate 'context' column
task1_1data['input'] = rawdata_3['parse1']+ " (영문 품명 : "+ rawdata_3['세부영문품명']+ ")"+rawdata_3['parse2']


# Generate 'response' column
task1_1data['output'] ="품명 : " + rawdata_3['품명'] + " (영문 품명 : "+ rawdata_3['영문품명']+ ") 에 해당합니다."
task1_1data['instruction'] = "다음은 상품의 세부품목, 제조업체명, 상품모델명에 대한 설명입니다. 이 상품이 해당되는 품명을 예측하시오."
task1_1data['category'] = "task1"



task1_1data = task1_1data[['instruction','input','output']]

task1_1data.head()
```

```python
#instruction #context #response #category

task1_2data = pd.DataFrame()

# Generate 'context' column
task1_2data['input'] = rawdata_4['parse1']+ " (영문 품명 : "+ rawdata_4['세부영문품명']+ ")"+rawdata_4['parse2']


# Generate 'response' column
task1_2data['output'] ="품명 : " + rawdata_4['품명'] + " (영문 품명 : "+ rawdata_4['영문품명']+ ") 에 해당합니다."
task1_2data['instruction'] = "다음은 상품의 세부품목, 제조업체명, 상품모델명에 대한 설명입니다. 이 상품이 해당되는 품명을 예측하시오."
task1_2data['category'] = "task1"



task1_2data = task1_2data[['instruction','input','output']]

task1_2data.head()
```

```python
unsp = pd.read_excel(file_path + '/KISTI/rawdata/UNv240301.xlsx',skiprows=12)
```

```python
rawdata= pd.read_csv(file_path + '/KISTI/rawdata/조달청 분류 체계 전체자료.csv')
```

```python
rawdata8 = rawdata[rawdata['thngClsfcNo'].between(10000000,99999999)]
rawdata6 = rawdata[rawdata['thngClsfcNo'].between(100000,999999)] #unspsc class 에 해당
rawdata4 = rawdata[rawdata['thngClsfcNo'].between(1000,9999)] #unspsc class 에 해당
rawdata2 = rawdata[rawdata['thngClsfcNo'].between(10,99)] #unspsc class 에 해당
```

```python
unsp2 = unsp[['Segment', 'Segment Title', 'Segment Definition']]
unsp4 = unsp[['Family', 'Family Title', 'Family Definition']]
unsp6 = unsp[['Class', 'Class Title', 'Class Definition']]
unsp8 = unsp[['Commodity', 'Commodity Title', 'Commodity Definition']]
```

```python
unsp8 = unsp8.dropna(subset=['Commodity'])  # 8자리 분류체계 명 NaN 값을 제거
unsp8 = unsp8.dropna(subset=['Commodity Definition']) # 8자리 분류체계 부연설명 NaN값 제거
unsp8['Commodity'] = unsp8['Commodity'].astype(int)  # 데이터 타입 변경
merged8 = pd.merge(rawdata8, unsp8, left_on='thngClsfcNo', right_on='Commodity')
```

```python
unsp6 = unsp6.dropna(subset=['Class', 'Class Title', 'Class Definition'])  # 8자리 분류체계 명 NaN 값을 제거
unsp6 = unsp6.drop_duplicates(subset=['Class'])
unsp6['Class'] = unsp6['Class'].astype(int)  # 데이터 타입 변경
unsp6['Class'] = unsp6['Class'].astype(str).str[:-2].str.zfill(6).astype(int)
merged6 = pd.merge(rawdata6, unsp6, left_on='thngClsfcNo', right_on='Class')
```

```python
unsp4 = unsp4.dropna(subset=['Family', 'Family Title', 'Family Definition'])  # 8자리 분류체계 명 NaN 값을 제거
unsp4 = unsp4.drop_duplicates(subset=['Family'])
unsp4['Family'] = unsp4['Family'].astype(int)  # 데이터 타입 변경
unsp4['Family'] = unsp4['Family'].astype(str).str[:-4].str.zfill(4).astype(int)
merged4 = pd.merge(rawdata4, unsp4, left_on='thngClsfcNo', right_on='Family')
```

```python
unsp2 = unsp2.dropna(subset=['Segment', 'Segment Title', 'Segment Definition'])  # 8자리 분류체계 명 NaN 값을 제거
unsp2 = unsp2.drop_duplicates(subset=['Segment'])
unsp2['Segment'] = unsp2['Segment'].astype(int)  # 데이터 타입 변경
unsp2['Segment'] = unsp2['Segment'].astype(str).str[:-6].str.zfill(6).astype(int)
merged2 = pd.merge(rawdata2, unsp2, left_on='thngClsfcNo', right_on='Segment')
```

```python
print(len(merged8))
print(len(rawdata8))
print(len(merged6))
print(len(rawdata6))
print(len(merged4))
print(len(rawdata4))
print(len(merged2))
print(len(rawdata2))
print(len(pd.concat([rawdata2,rawdata4,rawdata6,rawdata8])))
print(len(pd.concat([merged2,merged4,merged6,merged8])))
```

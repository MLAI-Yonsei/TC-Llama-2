### 예시  
**instruction**  


**input**  
다음 상품을 계층적으로 4단계로 분류하였을 때,
level 1 : 생활/주방/건강 , level 2 : 마스크/의약외품 , level 3 : 마스크, level 4 : 일반마스크 카테고리로 분류할 수 있습니다.
                
**output**  


```python
#instruction #input #output

data = pd.DataFrame()

# Generate 'context' column
data['input'] = "다음 상품을 계층적으로 4단계로 분류하였을 때, "+'level 1: ' + rawdata['LEVEL1'] \
                        + ' level 2: ' + rawdata['LEVEL2'] \
                        + ' level 3: ' + rawdata['LEVEL3'] \
                        + ' level 4: ' + rawdata['LEVEL4'] + " 카테고리로 분류할 수 있습니다."

# Generate 'response' column
data['output'] =""
data['instruction'] = ""


data = data[['instruction','input','output']]

data.head()
```

```python
danawa_category_csv = pd.read_csv(file_path + "/KISTI/task12_generating/danawa_category.csv")
```

```python
danawa_category_csv
```

```python
# 데이터프레임을 딕셔너리로 변환하여 저장할 리스트 생성
data_list1 = []

# 데이터프레임의 각 행에 대해 딕셔너리 생성 후 리스트에 추가
for _, row in danawa_category_csv.iterrows():
    data_dict1 = {
        'instruction' : "",
        'input': row['input'] ,
        'output': ""
    }
    data_list1.append(data_dict1)
```

```python
#딕셔너리 리스트를 JSON파일로 저장
with open(file_path + '/KISTI/task12_generating/danawa_category.json', 'w',encoding='utf-8') as f:
    json.dump(data_list1, f, ensure_ascii=False)
```

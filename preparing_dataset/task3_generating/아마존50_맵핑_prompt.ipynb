```python
amazon50_csv = pd.read_csv(file_path + "/KISTI/task3_generating/amazon50.csv")
```

```python
amazon50_csv
```

```python
# 데이터프레임을 딕셔너리로 변환하여 저장할 리스트 생성
data_list1 = []

# 데이터프레임의 각 행에 대해 딕셔너리 생성 후 리스트에 추가
for _, row in amazon50_csv.iterrows():
    data_dict1 = {
        'instruction' : row['instruction'],
        'input': row['input'] ,
        'output': row['output']
    }
    data_list1.append(data_dict1)
```

```python
#딕셔너리 리스트를 JSON파일로 저장
with open(file_path + "/KISTI/task3_generating/amazon50.json", 'w',encoding='utf-8') as f:
    json.dump(data_list1, f, ensure_ascii=False)
```

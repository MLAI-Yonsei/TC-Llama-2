```python
#Title 열 정보 중, 분류, 품목, 종류, 제품분류 정보를 제외하고 사용하려함(해당 데이터는 카테고리분류와 중복되므로)
keywords = ["분류", "품목", "종류", "제품분류"]
rawdata1['title'] = rawdata1['title'].fillna('nan').astype(str)
#title열에서 키워드를 포함하지 않는 행들 필터링해서 detailss 프레임으로 저장
detailss = rawdata1[~rawdata1['title'].apply(lambda x: any(keyword in x for keyword in keywords))]
```

```python
#detailss 데이터프레임의 title,content 열 합쳐서 새로운 열 생성
detailss['title_content'] = detailss['title'].astype(str)+" : "+detailss["content"].astype(str)

detailss.head()
```

```python
#detialss 데이터프레임으로 pcode 기준으로 그룹화하여 title_content 열의 값들을 결합
detailss2 = detailss.groupby('pcode')['title_content'].apply(', '.join)
```

```python
detailss2.head()
```

```python
#detailss2로 만든 데이터프레임을 50개 매칭한 rawdata와 pcode 기준으로 병합
detailss3 = pd.merge(rawdata,detailss2, on='pcode')
detailss3 = detailss3.drop_duplicates(['pcode'])
detailss3.head(20)
```

```python
#rawdata 를 새로운 데이터프레임으로 지정
rawdata = detailss3
```

**input**  
이 상품의 이름은 코멧 일회용 마스크 대형 화이트 100개입 (100개) 이고, 해당 제품의 상세 설명은
제조회사 : 쿠팡 코멧, 등록년월 : 2022년 09월, 크기 : 대형, 색상 : 화, 형태 : 덴탈형, 여름용 : O, 미세먼지차단 : O,
필터구성 : 3중구, 포장 : 묶음포장, 수량 : 100개, 적합성평가인증 : 상세설명 / 판매 사이트 문의 인증번호 확인,
안전확인인증 : 상세설명 / 판매 사이트 문의 인증번호 확인 .
다음 상품을 계층적으로 4단계로 분류하였을 때, level 1 : 생활/주방/건강 , level 2 : 마스크/의약외품 , level 3 : 마스크,
level 4 : 일반마스크 카테고리로 분류할 수 있습니다.

**output**  
다음 물품 분류는 보건용마스크, 영어로 Health masks 입니다.
보건용마스크의 상세설명은 다음과 같습니다. 호흡기를 질병의 감염, 매연으로부터 보호할 목적으로 사용하는 마스크. [식약처분류:32200]

```python
#instruction #input #output

data = pd.DataFrame()

# Generate 'context' column
data['input'] = "이 상품의 이름은 " + rawdata['다나와상품명'] +"이고, 해당 상품의 상세 설명은 " \
                + rawdata['title_content']+ "\n"+" 다음 상품을 계층적으로 4단계로 분류하였을 때, level 1: "  + rawdata['다나와 level1']\
                        + ' level 2: ' + rawdata['다나와 level2']\
                        + ' level 3: ' + rawdata['다나와 level3'] \
                        + ' level 4: ' + rawdata['다나와 level4'] + " 카테고리로 분류할 수 있습니다."


# Generate 'response' column
data['output'] ="다음 물품 분류는 " + rawdata['조달청분류체계 Nm'] +", 영어로 " + rawdata['조달청분류체계 engNm'] +" 입니다. " \
                + rawdata['조달청분류체계 Nm']+"의 상세설명은 다음과 같습니다. " + rawdata["조달청분류체계 Dscrpt"]


data['instruction'] = ""


data = data[['instruction','input','output']]

data.head()
```

```python
danawa50_csv = pd.read_csv(file_path + "/KISTI/task3_generating/danawa50.csv")
```

```python
danawa50_csv
```

```python
# 데이터프레임을 딕셔너리로 변환하여 저장할 리스트 생성
data_list1 = []

# 데이터프레임의 각 행에 대해 딕셔너리 생성 후 리스트에 추가
for _, row in danawa50_csv.iterrows():
    data_dict1 = {
        'instruction' : "",
        'input': row['input'] ,
        'output': row['output']
    }
    data_list1.append(data_dict1)
```

```python
#딕셔너리 리스트를 JSON파일로 저장
with open(file_path + "/KISTI/task3_generating/danawa50.json", 'w',encoding='utf-8') as f:
    json.dump(data_list1, f, ensure_ascii=False)
```
